```python
#converting column names into lower cases, so need not to worry about case sensitivity
df.columns=df.columns.str.strip().str.lower()
```

```python
df.info()
```

```python
df.head()
```

```python
#Standardizing name column
#print("before cleaning")
#print(df.name.unique())
df.name= df.name.str.lower()
#print("after cleaning")
#print(df.name.unique())
```

```python
df.head()
```

```python
#Filling null values with 0
df.isnull().sum()
df.age=df.age.fillna(0)
df.cabin=df.cabin.fillna(0)
df.embarked=df.embarked.fillna(0)
df.isnull().sum()
```

```python
#Display the count of different values in specific column
sb.countplot(x='sex',data=df)
plt.title("Sex")
plt.show()
# The below plot clearly shows that the count of male is more compared to female
```

```python
sb.countplot(x='sex',hue='pclass',data=df)
plt.title("gender based on class")
plt.show()
#Male and female population is more in class that other two classes
```

```python
sb.countplot(x='survived',hue='embarked',data=df)
plt.title("Survival based on embark")
plt.show()
#Below plot shows the number of survival and those not survived. It clearly shows that the survivers are less and 
#among those who survived are embarked from S
```
